```python
def forward(meas_seq, T, E):
    num_meas = len(meas_seq)
    N = T.shape[0]
    alpha = np.zeros((num_meas, N))
    pi = [0.25, 0.25, 0.25, 0.25]
    alpha[0] = pi*E[:,meas_seq[0]]
    for t in range(1, num_meas):
        alpha[t] = alpha[t-1].dot(T) * E[:, meas_seq[t]]
    return alpha
```
