```python
import qutip as qu
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dim = 50  # cut-off Dimension (Take care!)
```

For simplicity, $H_0 = \Lambda_1 \hat a^\dagger + h.c.$ (18MHz)

```python
a_op = qu.destroy(Dim)  # annihilation op.
Lambda_1 = 1  # Driving strength
Drive_op = Lambda_1 * a_op.dag()
H0 = Drive_op + Drive_op.dag()  # Hamiltonian
```

Consider 1-photon loss error: $\kappa \mathcal{D} [\hat a]$, where $\kappa \mathcal{D} [\hat a]\rho = \kappa (\hat a \rho \hat a^\dagger - \frac{1}{2}\{\hat a^\dagger \hat a, \rho\})$. $\kappa = \frac{1}{T_1}$.

```python
kappa = 1 / 180
n_th = 0.001
kappa_phi = 0
dissi_op = [np.sqrt(kappa * (1 + n_th)) * a_op, np.sqrt(kappa * (n_th)) * a_op.dag(), 
            np.sqrt(kappa_phi) * a_op.dag() * a_op]
```

```python
chi = 0.01/180
H_Kerr = chi / 2 * a_op.dag() * a_op.dag() * a_op * a_op
H_total = H0 + H_Kerr
```

Compute the evolution that start at $|n\rangle$ and the probability that detect the $|n+1\rangle$ state at time $t$.

```python
t_list = np.linspace(0,1,1000)
for n0 in range(4):
    state_init = qu.basis(Dim,n0)
    state_final = qu.basis(Dim, n0 + 1)
    e_op = [qu.ket2dm(state_final)]  # compute the expectation value with this operator
    result = qu.mesolve(H_total, state_init, t_list, dissi_op, e_op)
    plt.plot(t_list ** 2, result.expect[0], label=str(n0)+" to "+str(n0+1))
plt.legend()
plt.ylim(1E-5,0.5)
plt.xlim(1E-5,1)
plt.xscale('log')
plt.yscale('log')
```
