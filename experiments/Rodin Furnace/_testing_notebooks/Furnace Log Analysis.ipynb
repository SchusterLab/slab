```python
import slab
from slab import *
import h5py
from h5py import File
from matplotlib import pyplot as plt
os.chdir('S:\\_Furnace\\Furnace_Data\\')
%matplotlib notebook
```

```python
fnum='00005'
prefix='_2019-06-07_Furnace_log'
suffix='.h5'

with File(fnum+prefix+suffix, 'r') as f:
    tpts=np.array(f['tpts']-f['tpts'][0])
    volts=np.array(f['volts'])
    curr=np.array(f['curr'])
    pp_vals={}
    keys = [key for key in f.keys()]
    for vals in keys:
        if vals!='tpts' and vals!='volts' and vals!='curr' and vals!='fil_state':
            pp_vals[vals]=np.array(f[vals])
            

fig, ax=plt.subplots(figsize=(10,6))
for key in iter(pp_vals.keys()):
    plt.semilogy(tpts, pp_vals[key], label=key)
ax.set_ylim([5E-11,1E-2])

ax.set_xlabel('Elapsed Time (Seconds)')
ax.set_ylabel('Pressure (mbar)')

ax2 = ax.twinx()


plt.title('Pressure and Applied Power vs Time log: %s'%fnum+prefix)

ax2.plot(tpts, volts*curr, '-c', linewidth=3, label='Measured Coil Power')

ax2.set_ylabel('Applied Power (Watts)', color='c')
ax2.tick_params(axis='y', labelcolor='c')
ax.legend(loc=2)
plt.xlim([0,tpts[-1]+100])
plt.savefig(fnum+prefix+".png", bbox_inches='tight',dpi=500)
```
